```python
#3A11
try:
    a, b, c = map(int, input("Nhập độ dài 3 cạnh của tam giác").split())
    if (a + b > c) and (a + c > b) and (b + c > a):
        if a == b or b == c or a == c:
            if (a**2 + b**2 == c**2) or (b**2 + c**2 == a**2) or (a**2 + c**2 == b**2):
                print("Tam giác vuông cân")
            elif a == b and b == c:
                print("Tam giác đều")
            else:
                print("Tam giác cân")
        elif (a**2 + b**2 == c**2) or (b**2 + c**2 == a**2) or (a**2 + c**2 == b**2):
            print("Tam giác vuông")
        else:
            print("Tam giác thường")
    else:
        print("Không phải tam giác")
except ValueError:
    print("Giá trị không hợp lệ")
```

```python
#3A16
try:
    n = float(input("Nhập một số thực"))
    if n >= 0:
        du_duong = n % 1
        up_duong = n - du_duong + 1
        down_duong = n - du_duong
        if du_duong >= 0.5:
            nguyen_duong = up_duong
        else:
            nguyen_duong = down_duong
        print(f"Kết quả làm tròn lên, làm tròn xuống và làm tròn tới số nguyên gần nhất của {n} lần lượt là: {up_duong} {down_duong} {nguyen_duong}")
    else:
        du_am = -1*(n % -1)
        down_am = n + du_am -1
        up_am = n + du_am
        if du_am <= 0.5:
            nguyen_am = up_am
        else:
            nguyen_am = down_am
        print(f"Kết quả làm tròn lên, làm tròn xuống và làm tròn tới số nguyên gần nhất của {n} lần lượt là: {up_am} {down_am} {nguyen_am}")
except ValueError:
    print("Giá trị bạn nhập không hợp lệ")
```
